The aim of this notebook is to explore data about Airbnb listings in the city of Porto, Portugal. Exploring this data, I seek to understand pricing, availability, revenue and guest review patterns.

The columns that I intend to convert to integer values need a bit more investigation, as we can't have integer columns with missing values.  
In the column `bedrooms` I decided to calculate the mean of the column and replace the missing values by that value, as we can assume that all listings have at least one bedroom, and using the mean as a replacement would mean the average of bedrooms in our dataset won't be altered.  
To note the fact that the column `bathroom_text` is currently styled as "(number) baths (or other text)". This column is describing bathrooms with just a toilet as a 0.5 bathroom, but in cases where there are two of those, we wouldn't be aware, as it would sum to a full number, so I'm going to convert the column to integer type by removing the text and rounding up to the nearest integer so as to give us the total number of bathrooms that we are aware of.  
The column `price` will be converted to float so we can use it to do calculations, but before I can do that, I need to remove the "$" sign from that column.

```python
# Investigating the 'bedrooms' column
display(listings['bedrooms'].mean())
```

The mean of bedrooms per listing in our database is 1, so I'm going to replace the missing values in the column `bedrooms` by the number 1.  
Following that, I'll drop from the dataframe all rows where the `price` value is empty, and then I will convert `bathroom_text` to integer type by removing the numeric part and changing the type of the variable to `int`. 

```python
# Removing the lines where 'price' is empty
listings.dropna(subset=['price'], inplace =True)

# Replacing values in ´host_listings_count´ and ´bedrooms´
listings['bedrooms'] = listings['bedrooms'].fillna(1)

print('Bathrooms before:')
display(listings['bathrooms_text'].head(3))

# Replacing ´bathroom_text´ by the value only
def clean_bathrooms(text):
    # Extract numeric part
    try:
        num = float(text.split()[0])
    except:
        num = 0
    # Round up to nearest integer
    return int(math.ceil(num))

# Apply function
listings['bathrooms_text'] = listings['bathrooms_text'].apply(clean_bathrooms)

print('\nBathrooms after:')
display(listings['bathrooms_text'].head(3))

print('\nPrice before:')
display(listings['price'].head(3))

# Removing the '$' sign on the column ´price´
listings['price'] = listings['price'].str.replace('$','')
listings['price'] = listings['price'].str.replace(',','')

print('\nPrice after:')
display(listings['price'].head(3))
```

Finally, I'll convert all other columns that need to change the variable typing.

```python
# Convering the necessary columns
listings[['id', 'host_id']] = listings[['id', 'host_id']].astype('category')
listings['bedrooms'] = listings['bedrooms'].astype('int64')
listings['bathrooms_text'] = listings['bathrooms_text'].astype('int64')
listings['price'] = listings['price'].astype('float64')
```

Since we converted all column typings to the ones desired and took care of the empty values that were giving us cause for concern, I'll rename the following four columns to make their names more meaningful and easier to understand:

```python
# Renaming the columns
listings= listings.rename(columns={'neighbourhood_group_cleansed': 'area', 'bathrooms_text': 'bathrooms', 'number_reviews_ltm': 'number_reviews_l365'})

# Previewing the data to confirm it's ready for analysis
listings.head()
```

#### 3.1 Listings

To understand better what composes the listings, I'll start by checking the average number of listings per host and what the maximum is to check if there is a big discrepancy and dispersion of the data.

```python
# Creating a dataframe with the number of listings per host
listing_host = listings.groupby('host_id', observed=True)['id'].count()

# Discovering the average and maximum number of listings per host
print('Average number of listings per host:',listing_host.mean().round(2), '\n\
Maximum number of listings from the same host:', listing_host.max())
```

Even though hosts have on average around 3 listings, which is perfectly reasonable, there seem to be some hosts with a very high number of listings. Let's confirm by printing the 10 hosts with the highest number of listings and the count of hosts with 10 or fewer listings:

```python
print('Hosts with the highest number of listings:',listing_host.sort_values(ascending = False).head(10))
print('\nNumber of hosts with fewer listings:',listing_host.value_counts().head(10))
```

It can be observed that while there are some hosts with very high numbers of listings, including one with 380 listings, most users have 10 or fewer listings, with the majority of hosts having only one listing.

-------------------------

There are four room types in our dataframe: `Entire home/apt`, `Hotel room`, `Private room` and `Shared room`, and I'm interested in looking at the distribution of the listings by room type:

```python
# Creating a dataframe with the room type count
room_types = listings['room_type'].value_counts().sort_index()

fig1, ax = plt.subplots(figsize=(8,6))
room_types.plot(kind = 'bar', 
                color = ['red', 'blue', 'green', 'orange'])
plt.tick_params(axis = 'x', rotation=0)
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

This plot shows us that the vast majority of listings offer entire apartments, accounting for almost 6 times more than the number of private rooms, which appear in second place, and then, even private rooms have a much higher number of listings than shared rooms or hotel rooms. This shows some possible preference from the user side demanding more of those types of properties to rent, which in turn allows for hosts to offer more of those room types.

----------

I'll now look into the occupancy rate for each room type. This means that for each room type I'll plot the number of listings that were occupied from the total number of listings for that room type.

```python
# Calculating the occupancy for the last year by room type
occupied_days = listings.groupby('room_type')['estimated_occupancy_l365d'].sum()

# Calculating total possible nigths per room type
possible_days = room_types * 365

# Calculating the occupancy rate
occupancy_rate = occupied_days / possible_days

fig2, ax = plt.subplots(figsize=(8,6))
ax = occupancy_rate.plot(kind = 'bar', 
                         color = ['red', 'blue', 'green', 'orange'])
plt.tick_params(axis = 'x', rotation=0)
plt.title('Occupancy Rate by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Occupancy Rate')

# Add values above bars
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.2f}",
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom', size = 8
    )
    
plt.show()
```

It's visible that even though the ´Entire home/apt´ room type has a vastly superior number of listings when compared to the other categories, only around a quarter of their availability was filled, while ´Private room' and ´Shared room´ both had a little over 15% of their occupancy filled. This means that while there are a lot more apartments or entire houses listed, the amount that actually get occupied is not that different from the listings of other categories.

---

Now I'll investigate the average revenue by room type and compare it among the different categories.

```python
# Calculating the average revenue by room type
revenue_room_type = listings.groupby('room_type')['estimated_revenue_l365d'].mean()

fig3, ax = plt.subplots(figsize=(8, 6))
ax = revenue_room_type.plot(kind = 'bar', 
                            color = ['red', 'blue', 'green', 'orange'])
plt.tick_params(axis = 'x', rotation=0)
plt.title('Revenue by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Revenue')

# Add values above bars
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.0f}",
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom', size = 8
    )
    
plt.show()
```

When hosting entire homes or apartments, the overall average revenue is much bigger than the other listing types. Private rooms have the second-highest annual revenue at almost a third of the entire home, with shared rooms coming in third place with €1467 of average annual revenue. This is expected, as renting an entire house is much more expensive than independent rooms, and even within the different categories, the prices change with amenities provided.

---

Finaly, I'm going to compare how the number of bathrooms and bedrooms in a listing affect their average price.

```python
# creating dataframes to allow to plot avg price by bedroom and bathroom
bathrooms_price = listings.groupby('bathrooms')['price'].mean()
bedrooms_price = listings.groupby('bedrooms')['price'].mean()

fig4, ax = plt.subplots(1,2,figsize=(16,6))

bathrooms_price.plot(kind = 'bar', ax = ax[0])
ax[0].set_title('Average Price by Number of Bathrooms')
ax[0].set_ylabel('Average Price')
ax[0].tick_params(axis='x', rotation=0, labelsize = 'small')
ax[0].set_xlabel('Bathrooms')

bedrooms_price.plot(kind = 'bar')
ax[1].set_title('Average Price by Number of Bedrooms')
ax[1].set_ylabel('Average Price')
ax[1].tick_params(axis='x', rotation=0, labelsize = 'small')#
ax[1].set_xlabel('Bedrooms')

plt.tight_layout()
plt.show()
```

The number of bathrooms in a listing is directly proportional to the average price of listings, but the same can't be said about the number of bedrooms, which follows the same pattern until about 6 rooms, but then the price varies a lot, showing that there might be other factors to consider, such as the size of the rooms, view or other services offered, such as WiFi, jacuzzi, etc.

#### 3.2 Area

I'll now proceed to analyze information about the areas where the listings are located.  
I'll start by plotting the distribution of the number of listings per area:

```python
# Calculating variables to plot
area_listings = listings['area'].value_counts().sort_index()
# Get unique room types and areas
room_types = listings['room_type'].unique()
areas = listings['area'].unique()
areas.sort()  # Sort areas for consistent ordering

# Create figure with 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Number of Listings per Area by Room Type", fontsize=16)

# Define colormap - one color per area
colors = plt.cm.tab20(np.linspace(0, 1, len(areas)))
# Create a color dictionary for consistent colors
color_dict = {area: colors[i] for i, area in enumerate(areas)}

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Create a subplot for each room type
# Your previous code up to the loop
for i, room_type in enumerate(room_types):
    if i >= len(axes):  # In case there are more room types than subplots
        break
        
    # Filter data for this room type
    room_data = listings[listings['room_type'] == room_type]
    # Count listings by area for this room type
    area_count = room_data['area'].value_counts()
    
    # Ensure all areas are represented (even with zero count)
    for area in areas:
        if area not in area_count.index:
            area_count[area] = 0
    area_count = area_count.reindex(areas)  # Reindex to ensure consistent order
    
    # Get colors for each bar based on area
    bar_colors = [color_dict[area] for area in area_count.index]
    
    # Plot bar chart
    axes[i].bar(area_count.index, area_count.values, color=bar_colors)
    axes[i].set_title(f"Room Type: {room_type}")
    axes[i].set_ylabel("Count")
    
    # For top row subplots (even indices), hide x-tick labels
    if i == 0 or i == 1:
        # Set empty labels instead of empty ticks
        axes[i].set_xticklabels([])
        axes[i].set_xlabel("Area")
    else:
        # For bottom row, rotate labels
        axes[i].tick_params(axis='x', rotation=45)
        # Aligning the labels so they are readable
        for label in axes[i].get_xticklabels():
            label.set_ha('right')


# Hide any unused subplots
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

# Add a legend to show area colors (in the last subplot or a separate one)
if len(room_types) < len(axes):
    # Use the last subplot for the legend
    legend_ax = axes[-1]
    legend_ax.set_visible(True)
    legend_ax.axis('off')
    
    # Create legend handles
    handles = [plt.Rectangle((0,0),1,1, color=color_dict[area]) for area in areas]
    legend_ax.legend(handles, areas, loc='center', title="Areas")


plt.tight_layout()
plt.subplots_adjust(top=0.92)  # Make room for the suptitle
plt.show()
```

Central Porto is the area with the most listings by far in all types of rooms, the next two areas with most listings are `Vila do Conde` and `Vila Nova de Gaia`.

I'll now plot the maximum and minimum prices for each area to see the discrepancy between those two values.

```python
# Calculate average price by area and room type
max_price_area = listings.groupby(['area', 'room_type'])['price'].max().sort_index()
min_price_area = listings.groupby(['area', 'room_type'])['price'].min().sort_index()

# Get unique areas and room types
areas = max_price_area.index.get_level_values('area').unique()
room_types = max_price_area.index.get_level_values('room_type').unique()

# Create a color map for areas
import matplotlib.cm as cm
area_colors = cm.tab20(np.linspace(0, 1, len(areas)))
area_color_map = dict(zip(areas, area_colors))

# Create a figure with 4 subplots (one for each room type)
fig6, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()  # Flatten to make it easier to iterate

# Plot each room type in a separate subplot
for i, room_type in enumerate(room_types):
    # Filter data for this room type
    room_data_max = max_price_area.xs(room_type, level='room_type')
    room_data_min = min_price_area.xs(room_type, level='room_type')
    
    # Create bar chart
    bars = axes[i].bar(room_data_max.index, room_data_max.values, 
                       color=[area_color_map[area] for area in room_data_max.index], alpha = 0.4)
    bars2 = axes[i].bar(room_data_min.index, room_data_min.values, 
                       color=[area_color_map[area] for area in room_data_min.index], alpha = 0.6)
    
    # Add title and labels
    axes[i].set_title(f'Prices by Area for {room_type}')
    axes[i].set_xlabel('Area')
    axes[i].set_ylabel('Price')
    
    # Rotate x-axis labels for better readability
    axes[i].tick_params(axis='x', rotation=45, labelsize='small')
    for label in axes[i].get_xticklabels():
        label.set_ha('right')
 
# Create a legend for areas
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=color, label=area) 
                  for area, color in area_color_map.items()]

# Add the legend to the figure
fig.legend(handles=legend_elements, title="Areas", 
           loc='lower center', bbox_to_anchor=(0.5, 0.02), ncol=len(areas)//2)

plt.tight_layout()
plt.subplots_adjust(bottom=0.15)  # Make room for the legend
plt.show()
```

There is such a big discrepancy between the maximum and minimum prices, especially for `Entire home/Apt` and `Private room`, that the minimum values are barely visible on the plots. Those differences between the two values might be due to seasonality, other amenities, and the level of comfort offered by the listing. The type of room also has a big impact on the price of the venue, as shared rooms will be much cheaper than entire houses.

#### 3.3 Availability

Next, I'll look into the number of available days in the next few months, for the three areas with the most listings.

```python
# Creating dataframe with the 3 top listing areas
top3 = listings['area'].value_counts().nlargest(3).index
top3_listings = listings[listings['area'].isin(top3)]

availability_cols = ['availability_30', 'availability_60', 'availability_90']

fig7, ax = plt.subplots(3, 1, figsize=(10, 9), sharex=False)

plt.suptitle('Future availability for the 3 top listed areas')

for i, col in enumerate(availability_cols):
    for area in top3:
        subset = top3_listings[top3_listings['area'] == area]
        ax[i].hist(subset[col], 
                   bins = (i+1)*30, 
                   alpha=0.5, 
                   label=area)

    ax[i].set_xlabel(f'Number of available days')
    ax[i].set_ylabel('Number of listings')
    ax[i].set_title(f'Number of available days in the next {col.split("_")[1]} days')
 
ax[0].legend()
plt.tight_layout()
plt.show()
```

In the availability count for the listings, the availability for the next month follows a right-skewed normal distribution. Most of the listings have around 5 free days for bookings in the next month, but this number decreases as the month progresses.  
In the next two months, most listings have around 15 to 45 days available, while in the next three months, most listings have around 45 to 80 days without occupancy.  
Of note is that a big number of listings in all cases report having no availability or having no bookings, as seen by the columns at the beginning and end of the three plots.
From the above plots, it's visible that as we start to look at a wider range of days, the more free days we have, meaning that dates closer to the actual day have fewer free spaces to rent.  

Next, I will look into the relationship between revenue and occupancy in the last year for each room type. I expect to see that the revenue increases as the occupancy increases for all room types, but that the most profitable room type is `Entire home`.

```python
fig8 = sns.lmplot(
    data=listings, 
    x='estimated_occupancy_l365d', 
    y='estimated_revenue_l365d',
    height=6,
    aspect=1.2,
    hue='room_type',
    scatter_kws={'alpha':0.4})
plt.title('Relation between Occupancy and Revenue in the last 365 days')
plt.xlabel('Occupancy')
plt.ylabel('Revenue')
ax = fig8.axes[0, 0]
# Remove the original legend
fig8._legend.remove()

# Create a new legend inside the plot
ax.legend(loc='upper right', title='Area')

plt.tight_layout()
plt.show()
```

As I expected, the most profitable types of accommodations for renting are entire homes or apartments. Even with a lower occupancy rate and taking out outliers with revenue way above the average, it's visible that this type of property has higher revenue than the rest. To note as well that the second most profitable room type is the `Private room`, and that as the occupancy increases, so does the revenue.

#### 3.5 Correlation Matrix

Below I'll plot the correlation matrix between all the variables. This matrix has values close to one for a strong direct correlation between the variables (as one variable increases so does the other) and values close to -1 for strong inverse correlation. Values close to zero mean the two variables don't appear to have a correlation among themselves.

```python
# First, select only numeric columns for correlation
numeric_columns = listings.select_dtypes(include=['float64', 'int64'])

fig10, ax = plt.subplots(figsize=(10,6))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm')

# Aligning the labels
ax.tick_params(axis='x', rotation=45, labelsize = 'small')
for label in ax.get_xticklabels():
    label.set_ha('right')

plt.title("Correlation Matrix")
```

Here, we can see the relationship between all the variables. The number of bedrooms and bathrooms has a medium-strong correlation, most likely due to the fact that `Entire home` and most `Private rooms` have at least a bathroom. As expected, the availabilities in the next few months are all related to each other, becoming weaker as the timespan is increased. Estimated occupancy and estimated revenue in the last year also have a direct correlation with medium strength, almost at 0.5. The review scores are all corrrelated with each other, with value and accuracy having a heavier impact on the overall rating. Intererstingly enough, `Price` seems to be very independant from any other variables on the database, with the strongest correlation at 0.34 with the estimated revenue in the last year.

On this project, I was amazed at the quantity of information available on the dataset I downloaded. Even though most of the columns weren't necessary for the analysis that I wanted to pursue, other types of analysis could be conducted with the same dataset, for example, examining the hosts more in-depth and the patterns that they might follow.

#### Key findings
 - Most listings are for entire homes, with more than 11000 listings in this category, followed by private rooms, with around 2000. This might be an example of demand and supply in this market.
 - Porto has a much higher number of listings than all other areas combined, but it's the sixth area with the lowest average price, while also being one of the areas with the highest maximum price.
 - The most profitable accommodations are `Entire house/Apt` at all occupancy levels, and as the number of occupied days increases, so does the revenue.
 - The price of a listing doesn't seem to be directly related to other variables in this dataset.

#### Other insights
 - Most hosts (3039) have only one listings, but there is also a high number of hosts with 2 (637) and 3 (348) listings.
 - There are some hosts with a high number of listings, most likely companies or hotels, including one with 380 listings!
 - 25% of entire house vacancies are occupied yearly, while for `Private room` and `Shared room` the value is about 15%.
 - As the number of bathrooms available increases, so does the average price, but for the bedrooms, this value is not so directly correlated. This might be due to the size of the bedroom, the view, or other commodities.
 - There is a huge discrepancy between the maximum and minimum prices across all areas, especially in `Entire house/Apt` and `Private room` which might be due to seasonality, view, size of property listed, or other commodities.
 - The closer the booked day is to the current day, the fewer available rooms there are, and as the period window increases, the available number of rooms does as well.
 - The peaks for average prices based on rating score are around 2.6 and 3.5, meaning that higher-priced venues are not scored higher than the other listings.
 - The availability of a month affects the number of free venues going forward, as was expected.
 - The overall score rating seems to be more influenced by the accuracy and value scores of the listing than the other scores.

#### Recommendations
 - As a host, it seems to be better to invest in renting entire houses if possible, as they have higher returns and occupancy rates, if not, the second-best choice would be private rooms.
 - As a renter, the area doesn't affect the price much, so it's better to plan your itinerary first and then search in the location close to it. If someone is looking for a cheaper option, renting private rooms is probably a better option.
