# Airbnb in Porto

## Data Analysis and Exploratory Data Insights

The aim of this notebook is to explore data about Airbnb listings in the city of Porto, Portugal. Exploring this data, I seek to understand pricing, availability, revenue and guest review patterns.


### Main Questions

 - What drives the prices of Airbnb listings in Porto?
 - Which neighbourhoods generate the highest estimated revenue?
 - How do room type, number of bedrooms and occupancy influence price?
 - How do review scores relate to listing performance?


#### Data source

The data was acquired through the website [InsideAirbnb](https://insideairbnb.com/get-the-data/).

### 1. Importing the data

To start, I'll import the libraries and modules that I'll be using through this notebook. Then, I'll import the `.csv` file into a dataframe so I can work upon it. After loading the dataframe, I'll confirm its integrety by showcasing the first five rows of it.

```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import math
```

```python
# Importing the data into a dataframe
listings = pd.read_csv('listings.csv')

# Previewing to make sure data was imported correctly
display(listings.head())
```

### 2. Data cleaning and preprocessing

Now that the dataframe is set up, it's time to do some data wrangling. I'll start by checking the data format and structure to transform it so that I can use it.  
I'll start by listing all the columns in the dataframe and checking its shape for the number of columns and rows.

```python
# Checking shape and columns of the dataframes
display(listings.columns)
print('listings size:', listings.shape)
```

The dataframe has an immense amount of data on it. It has 14,806 rows, one for each individual listing, and 79 columns (refer to the [attached file](Column_Description.xls)) with independent fields about that row's listing.  
Due to the amount of information, I'm going to drop several columns, keeping only the ones I find useful for my analysis: 

|`listings`|||||
|---|---|---|---|---|
|id|host_id|neighbourhood_group_cleansed|room_type|bathroom_text|
|bedrooms|price|availability_30|availability_60|availability_90|

```python
# Removing unnecessary columns
listings = listings.drop(columns=['listing_url', 'scrape_id', 'last_scraped', 'source', 'name', 'description', 'neighborhood_overview', 'picture_url',
                                  'host_url', 'host_name', 'host_location', 'host_about', 'host_response_time', 'host_response_rate', 
                                  'host_acceptance_rate', 'host_is_superhost', 'host_thumbnail_url', 'host_picture_url', 'host_neighbourhood', 
                                  'host_total_listings_count', 'host_verifications', 'host_has_profile_pic', 'host_identity_verified', 'neighbourhood',
                                  'latitude', 'longitude', 'property_type', 'bathrooms', 'beds', 'minimum_nights', 'maximum_nights', 
                                  'minimum_minimum_nights', 'maximum_minimum_nights', 'minimum_maximum_nights', 'maximum_maximum_nights', 
                                  'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm', 'calendar_updated', 'has_availability', 'calendar_last_scraped', 
                                  'availability_eoy', 'number_of_reviews_ly', 'first_review', 'last_review', 'license', 'instant_bookable', 
                                  'calculated_host_listings_count', 'reviews_per_month', 'host_since', 'host_listings_count', 'neighbourhood_cleansed',
                                  'accommodates', 'amenities', 'availability_365', 'number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d',
                                  'calculated_host_listings_count_entire_homes', 'calculated_host_listings_count_private_rooms',
                                  'calculated_host_listings_count_shared_rooms'])

# Confirming changes took effect
listings.head()
```

Now that the data that was of no use for this analysis was removed, I'll check the data types of each column to see if they match what would be expected and correct the typings if not. Then I'll check for missing values in the columns as well.

```python
print('Column types:')
display(listings.dtypes)
print('\nEmpty values:')
listings.isnull().sum()
```

Though there are some missing values on `bathroom_text`, `bedrooms`, `estimated_revenue_l365d` and all the `review_scores`, the most impactful column with missing values is `price` with 1614 missing values. For the first columns mentioned, I believe keeping the empty values won't impact the exploratory analysis, so I intend to keep them as empty, but for the `price` column, I believe keeping those values as empty, or even replacing the values for others such as `mean` or `mode` would impact possible calculations, so I made the decision to remove the missing price values.

As for the column types, we have:
|Column|Data type|Variable type|Change into|

The columns that I intend to convert to integer values need a bit more investigation, as we can't have integer columns with missing values.  
In the column `bedrooms` I decided to calculate the mean of the column and replace the missing values by that value, as we can assume that all listings have at least one bedroom, and using the mean as a replacement would mean the average of bedrooms in our dataset won't be altered.
To note the fact that the column `bathroom_text` is currently styled as "(number) baths (or other text)". This column is describing bathrooms with just a toilet as a 0.5 bathroom, but in cases where there are two of those, we wouldn't be aware, as it would sum to a full number, so I'm going to convert the column to integer type by removing the text and rounding up to the nearest integer so as to give us the total number of bathrooms that we are aware of.  
The column `price` will be converted to float so we can use it to do calculations, but before I can do that, I need to remove the "$" sign from that column.

```python
# Investigating the 'bedrooms' column
display(listings['bedrooms'].mean())
```

The mean of bedrooms per listing in our database is 1, so I'm going to replace the missing values in the column `bedrooms` by the number 1.  
Following that, I'll drop from the dataframe all rows where the `price` value is empty, and then I will convert `bathroom_text` to integer type by removing the numeric part and changing the type of the variable to `int`.  
Finally, I'll convert all other columns that need to change the variable typing.

```python
# Removing the lines where 'price' is empty
listings.dropna(subset=['price'], inplace =True)

# Replacing values in ´host_listings_count´ and ´bedrooms´
listings['bedrooms'] = listings['bedrooms'].fillna(1)

# Replacing ´bathroom_text´ by the value only
def clean_bathrooms(text):
    # Extract numeric part
    try:
        num = float(text.split()[0])
    except:
        num = 0
    # Round up to nearest integer
    return int(math.ceil(num))

# Apply function
listings['bathrooms_text'] = listings['bathrooms_text'].apply(clean_bathrooms)

# Removing the '$' sign on the column ´price´
listings['price'] = listings['price'].str.replace('$','')
listings['price'] = listings['price'].str.replace(',','')
```

```python
# Convering the necessary columns
listings[['id', 'host_id']] = listings[['id', 'host_id']].astype('category')
listings['bedrooms'] = listings['bedrooms'].astype('int64')
listings['bathrooms_text'] = listings['bathrooms_text'].astype('int64')
listings['price'] = listings['price'].astype('float64')
```

Since we converted all column typings to the ones desired and took care of the empty values that were giving us cause for concern, I'll rename the following four columns to make their names more meaningful and easier to understand:

```python
# Renaming the columns
listings= listings.rename(columns={'neighbourhood_group_cleansed': 'area', 'bathrooms_text': 'bathrooms', 'number_reviews_ltm': 'number_reviews_l365'})

# Previewing the data to confirm it's ready for analysis
listings.head()
```

### 3. Data Analysis

Now that the data is clean and sorted, we can start doing some exploration and visualisations.  
I'll start by exploring the listings themselves, such as the average number of listings per host and an exploration about the types of rooms, their occupancy rates and their revenue, etc. 
Secondly, I will be looking into the areas in the listings by comparing the average prices, number of listings and so on.  
After that, I'll explore the availability going forward, how it's spread by the number of unoccupied  days in the next few months, and what the relationship is between revenue and occupancy by room type.  
Finally, I'll explore how the ratings relate to the listings.

#### 3.1 Listings

To understand better what composes the listings, I'll start by checking the average number of listings per host and what the maximum is to check if there is a big discrepancy and dispersion of the data.

```python
# Creating a dataframe with the number of listings per host
listing_host = listings.groupby('host_id', observed=True)['id'].count()

# Discovering the average and maximum number of listings per host
print('Average number of listings per host:',listing_host.mean().round(2), '\n\
Maximum number of listings from the same host:', listing_host.max())
```

Even though hosts have on average around 3 listings, which is perfectly reasonable, there seem to be some hosts with a very high number of listings. Let's confirm by printing the 10 hosts with the highest number of listings and the count of hosts with 10 or fewer listings:

```python
print('Hosts with the highest number of listings:',listing_host.sort_values(ascending = False).head(10))
print('\nNumber of hosts with fewer listings:',listing_host.value_counts().head(10))
```

It can be observed that while there are some hosts with very high numbers of listings, including one with 380 listings, most users have 10 or fewer listings, with the majority of hosts having only one listing.

-------------------------

There are four room types in our dataframe: `Entire home/apt`, `Hotel room`, `Private room` and `Shared room`, and I'm interested in looking at the distribution of the listings by room type:

```python
# Creating a dataframe with the room type count
room_types = listings['room_type'].value_counts().sort_index()

fig1 = plt.figure(figsize=(5,3))
room_types.plot(kind = 'bar', 
                color = ['red', 'blue', 'green', 'orange'])
plt.tick_params(axis = 'x', rotation=0)
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

This plot shows us that the vast majority of listings offer entire apartments, accounting for almost 6 times more than the number of private rooms, which appear in second place, and then, even private rooms have a much higher number of listings than shared rooms or hotel rooms. This shows some possible preference from the user side demanding more of those types of properties to rent, which in turn allows for hosts to offer more of those room types.

----------

I'll now look into the occupancy rate for each room type. This means that for each room type I'll plot the number of listings that were occupied from the total number of listings for that room type.

```python
# Calculating the occupancy for the last year by room type
occupied_days = listings.groupby('room_type')['estimated_occupancy_l365d'].sum()

# Calculating total possible nigths per room type
possible_days = room_types * 365

# Calculating the occupancy rate
occupancy_rate = occupied_days / possible_days

fig2 = plt.figure(figsize=(5,3))
ax = occupancy_rate.plot(kind = 'bar', 
                         color = ['red', 'blue', 'green', 'orange'])
plt.tick_params(axis = 'x', rotation=0)
plt.title('Occupancy Rate by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Occupancy Rate')

# Add values above bars
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.2f}",
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom', size = 8
    )
    
plt.show()
```

It's visible that even though the ´Entire home/apt´ room type has a vastly superior number of listings when compared to the other categories, only around a quarter of their availability was filled, while ´Private room' and ´Shared room´ both had a little over 15% of their occupancy filled. This means that while there are a lot more apartments or entire houses listed, the amount that actually get occupied is not that different from the listings of other categories.

---

Now I'll investigate the average revenue by room type and compare it among the different categories.

```python
# Calculating the average revenue by room type
revenue_room_type = listings.groupby('room_type')['estimated_revenue_l365d'].mean()

fig3 = plt.figure(figsize=(5,3))
ax = revenue_room_type.plot(kind = 'bar', 
                            color = ['red', 'blue', 'green', 'orange'])
plt.tick_params(axis = 'x', rotation=0)
plt.title('Revenue by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Revenue')

# Add values above bars
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.0f}",
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom', size = 8
    )
    
plt.show()
```

When hosting entire homes or apartments, the overall average revenue is much bigger than the other listing types. Private rooms have the second-highest annual revenue at almost a third of the entire home, with shared rooms coming in third place with €1467 of average annual revenue. This is expected, as renting an entire house is much more expensive than independent rooms, and even within the different categories, the prices change with amenities provided.

---

Finaly, I'm going to compare how the number of bathrooms and bedrooms in a listing affect their average price.

```python
# creating dataframes to allow to plot avg price by bedroom and bathroom
bathrooms_price = listings.groupby('bathrooms')['price'].mean()
bedrooms_price = listings.groupby('bedrooms')['price'].mean()

fig4, ax = plt.subplots(1,2,figsize=(10,3))

bathrooms_price.plot(kind = 'bar', ax = ax[0])
ax[0].set_title('Average Price by Number of Bathrooms')
ax[0].set_ylabel('Average Price')

bedrooms_price.plot(kind = 'bar')
ax[1].set_title('Average Price by Number of Bedrooms')
ax[1].set_ylabel('Average Price')

plt.tight_layout()
plt.show()
```

The number of bathrooms in a listing is directly proportional to the average price of listings, but the same can't be said about the number of bedrooms, which follows the same pattern until about 6 rooms, but then the price varies a lot, showing that there might be other factors to consider, such as the size of the rooms, view or other services offered, such as WiFi, jacuzzi, etc.

#### 3.2 Area

```python
# Calculating variables to plot
area_listings = listings['area'].value_counts().sort_index()
price_area = listings.groupby('area')['price'].mean().sort_index()

# Defining colormap
colors = plt.cm.tab20(np.linspace(0, 1, len(area_listings)))

fig5, ax = plt.subplots(2,1,figsize = (10,6))
# Left plot – number of listings
ax[0].bar(area_listings.index, 
          area_listings.values, 
          color = colors)
ax[0].set_title("Number of Listings per Area")
ax[0].set_xlabel("Area")
ax[0].set_ylabel("Count")
ax[0].tick_params(axis='x', rotation=45, labelsize = 'small')
# Aligning the labels so they are readable
for label in ax[0].get_xticklabels():
    label.set_ha('right')

# Right plot – average price
ax[1].bar(price_area.index, 
          price_area.values, 
          color = colors)
ax[1].set_title("Average Listing Price per Area")
ax[1].set_xlabel("Area")
ax[1].set_ylabel("Average Price (€)")
ax[1].tick_params(axis='x', rotation=45, labelsize = 'small')
# Aligning the labels so they are readable
for label in ax[1].get_xticklabels():
    label.set_ha('right')

plt.tight_layout()
plt.show()
```

Central Porto is the area with the most listings by far, but when looking at the prices, it's actually one of the cheapest areas, while areas that are more expensive have fewer listings, like `Oliveira de Azemeis` or `Vale de Cambra`, probably because those are more secluded areas with a higher level of commodities and comfort.

```python
# making dataframes for maximum and minimum prices for comparasion
max_price_area = listings.groupby('area')['price'].max().sort_index()
min_price_area = listings.groupby('area')['price'].min().sort_index()

fig6, ax = plt.subplots(2,1,figsize = (10,6), sharex= True)
ax[0].bar(max_price_area.index, 
          max_price_area.values, 
          color = colors)
ax[1].bar(min_price_area.index, 
          min_price_area.values, 
          color = colors)

# Adding titles
ax[0].set_title('Maximum Average Price per Area')
ax[1].set_title('Minimum Average Price per Area')

# Aligning the labels so they are readable
ax[1].tick_params(axis='x', rotation=45, labelsize = 'small')
for label in ax[1].get_xticklabels():
    label.set_ha('right')

# Add values above bars
for p in ax[0].patches:
    ax[0].annotate(
        f"{p.get_height():.0f}",
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom', size = 8
    )
for p in ax[1].patches:
    ax[1].annotate(
        f"{p.get_height():.0f}",
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom', size = 8)
    
plt.tight_layout()
plt.show()
```

There is a big discrepancy in maximum and minimum prices, but that might be due to seasonality, area and other amenities and the level of comfort offered by the listing.

#### 3.3 Availability

```python
# Creating dataframe with the 3 top listing areas
top3 = listings['area'].value_counts().nlargest(3).index
top3_listings = listings[listings['area'].isin(top3)]

availability_cols = ['availability_30', 'availability_60', 'availability_90']

fig7, ax = plt.subplots(3, 1, figsize=(10, 8), sharex=False)

plt.suptitle('Future availability for the 3 top listed areas')

for i, col in enumerate(availability_cols):
    for area in top3:
        subset = top3_listings[top3_listings['area'] == area]
        ax[i].hist(subset[col], 
                   bins = (i+1)*30, 
                   alpha=0.5, 
                   label=area)

    ax[i].set_xlabel(f'Availability in the next {col.split("_")[1]} days')
    ax[i].set_ylabel('Number of listings')
    ax[i].set_title(f'Number of available nights in the next {col.split("_")[1]} days')
 
ax[0].legend()
plt.tight_layout()
plt.show()
```

In the availability count for the listings, the availability for the next month follows a right-skewed normal distribution. Most of the listings have around 5 free days for bookings in the next month, but this number decreases as the month progresses.  
In the next two months, most listings have around 15 to 45 days available, while in the next three months, most listings have around 45 to 80 days without occupancy.  
Of note is that a big number of listings in all cases report having no availability or having no bookings, as seen by the columns at the beginning and end of the three plots.

```python
fig8 = plt.figure(figsize=(10,6))

g=sns.lmplot(
    data=listings, 
    x='estimated_occupancy_l365d', 
    y='estimated_revenue_l365d',
    height=6,
    aspect=1.2,
    hue='room_type',
    scatter_kws={'alpha':0.4})
plt.title('Relation between Occupancy and Revenue in the last 365 days')
plt.xlabel('Occupancy')
plt.ylabel('Revenue')
ax = g.axes[0, 0]
# Remove the original legend
g._legend.remove()

# Create a new legend inside the plot
ax.legend(loc='upper right', title='Area')

plt.tight_layout()
plt.show()
```

The most profitable types of accommodations for renting are entire homes or apartments. Even with a lower occupancy rate and taking out outliers with revenue way above the average, it's visible that this type of property has higher revenue than the rest.

#### 3.4 Ratings

```python
fig9 = plt.figure(figsize=(10,6))

sns.lineplot(data = listings, x = 'review_scores_rating', y = 'price', errorbar= None)
sns.lineplot(data = listings, x = 'review_scores_accuracy', y = 'price', errorbar= None)
sns.lineplot(data = listings, x = 'review_scores_cleanliness', y = 'price', errorbar= None)
sns.lineplot(data = listings, x = 'review_scores_checkin', y = 'price', errorbar= None)
sns.lineplot(data = listings, x = 'review_scores_communication', y = 'price', errorbar= None)
sns.lineplot(data = listings, x = 'review_scores_location', y = 'price', errorbar= None)
sns.lineplot(data = listings, x = 'review_scores_value', y = 'price', errorbar= None)

# Titles
plt.title('Ratings Score')

# Labels
plt.xlabel('Score')
plt.ylabel('Price')

plt.legend(['Overall Rating','Accuracy Rating','Cleanliness Rating','Check-in Rating','Communication Rating','Location Rating','Value Rating'])

plt.tight_layout()
plt.show()
```

Above we can see that for ratings above 3.5, there are listings with a higher price that received good ratings for the accuracy of the listing, communication with the host and cleanliness, meaning that those factors can be factors that pull people towards a listing and that allows for an increase in price.
